```python
import shutil
import os
from tqdm import tqdm

def move_images(metadata, base_image_dir, output_base_dir):
    """
    Move images into train, val, and test directories based on metadata.

    Parameters:
        metadata (DataFrame): The metadata containing image paths and split info.
        base_image_dir (str): The base directory where original images are stored.
        output_base_dir (str): The base directory where images should be organized.
    """
    # Mapping TrainValTest values to subdirectories
    split_mapping = {1: "train", 2: "val", 3: "test"}

    for _, row in tqdm(metadata.iterrows(), desc="Moving images", total=len(metadata)):
        filename = row["Filename"]
        class_name = row["Class"]
        split = split_mapping[row["TrainValTest"]]

        # Source and destination paths
        source_path = os.path.join(base_image_dir, class_name, filename)
        dest_dir = os.path.join(output_base_dir, "images", split)
        dest_path = os.path.join(dest_dir, filename)

        # Ensure the source image exists
        if not os.path.exists(source_path):
            print(f"Warning: Source image not found: {source_path}")
            continue

        # Move the image
        shutil.copy2(source_path, dest_path)  # Use copy2 to preserve metadata
        #print(f"Moved: {source_path} -> {dest_path}")

# Move images for the original dataset
move_images(metadata, "./ExDark", "./datasets/ExDark")

# Move images for the preprocessed dataset
move_images(metadata, "./ExDark_Preprocessed", "./datasets/ExDark_Preprocessed")
```

```python
import os
import cv2
from tqdm import tqdm

def move_and_convert_annotations(metadata, annotation_dir, output_base_dir, base_image_dir):
    """
    Move and convert annotation files into YOLO format for train, val, and test directories based on metadata.

    Parameters:
        metadata (DataFrame): The metadata containing image paths and split info.
        annotation_dir (str): The directory where annotation files are stored (divided by class).
        output_base_dir (str): The base directory where annotations should be organized.
        base_image_dir (str): The directory containing all images (train, val, test combined).
    """
    # Mapping TrainValTest values to subdirectories
    split_mapping = {1: "train", 2: "val", 3: "test"}

    # Class mapping
    class_mapping = {
        "Bicycle": 0,
        "Boat": 1,
        "Bottle": 2,
        "Bus": 3,
        "Car": 4,
        "Cat": 5,
        "Chair": 6,
        "Cup": 7,
        "Dog": 8,
        "Motorbike": 9,
        "People": 10,
        "Table": 11
    }

    for _, row in tqdm(metadata.iterrows(), desc="Moving and converting annotations", total=len(metadata)):
        filename = row["Filename"]
        class_name = row["Class"]
        split = split_mapping[row["TrainValTest"]]

        # Path to the source annotation file (with image extension)
        annotation_filename = filename + ".txt"
        source_path = os.path.join(annotation_dir, class_name, annotation_filename)
        dest_dir = os.path.join(output_base_dir, "labels", split)

        # Construct destination path without the image extension (just .txt)
        dest_filename = filename.split('.')[0] + ".txt"  # Remove image extension (.jpg or .png)
        dest_path = os.path.join(dest_dir, dest_filename)

        # Ensure the source annotation exists; try multiple case variations
        if not os.path.exists(source_path):
            # Try lower-case suffixed annotation
            alternative_filename = filename.lower() + ".txt"
            source_path = os.path.join(annotation_dir, class_name, alternative_filename)

            if not os.path.exists(source_path):
                # Try upper-case suffixed annotation
                alternative_filename = filename.upper() + ".txt"
                source_path = os.path.join(annotation_dir, class_name, alternative_filename)

        # If still not found, skip the file
        if not os.path.exists(source_path):
            print(f"Warning: Annotation file not found: {source_path}")
            continue

        # Get the corresponding image path
        image_path = os.path.join(base_image_dir, split, filename)
        if not os.path.exists(image_path):
            print(f"Warning: Image not found for annotation: {image_path}")
            continue

        # Get image dimensions
        image = cv2.imread(image_path)
        if image is None:
            print(f"Error: Could not read image {image_path}")
            continue
        image_height, image_width, _ = image.shape  # Get height and width

        # Read and convert the annotation file
        with open(source_path, "r") as infile:
            yolo_annotations = []
            for line in infile:
                # Skip invalid lines or headers
                if line.startswith("%") or len(line.strip()) == 0:
                    continue

                parts = line.strip().split()
                original_class_name = parts[0]
                x, y, width, height = map(int, parts[1:5])

                # Map class name to class ID
                class_id = class_mapping.get(original_class_name, -1)
                if class_id == -1:
                    print(f"Warning: Class '{original_class_name}' not in class mapping. Skipping.")
                    continue

                # Convert to YOLO format (normalize bounding box)
                x_center = (x + width / 2) / image_width
                y_center = (y + height / 2) / image_height
                norm_width = width / image_width
                norm_height = height / image_height

                # Append the YOLO formatted annotation
                yolo_annotations.append(f"{class_id} {x_center:.6f} {y_center:.6f} {norm_width:.6f} {norm_height:.6f}")

        # Save the converted annotations to the destination path (without image extension)
        os.makedirs(dest_dir, exist_ok=True)
        with open(dest_path, "w") as outfile:
            outfile.write("\n".join(yolo_annotations))

# Move and convert annotations for the original dataset
move_and_convert_annotations(metadata, "./ExDark_Annno", "./datasets/ExDark", "./datasets/ExDark/images")

# Move and convert annotations for the preprocessed dataset
move_and_convert_annotations(metadata, "./ExDark_Annno", "./datasets/ExDark_Preprocessed", "./datasets/ExDark_Preprocessed/images")
```

```python
import matplotlib.pyplot as plt

def visualize_positive_predictions(results, num_images=5, conf_threshold=0.7):
    """
    Visualize high-confidence positive predictions from the test set.

    Parameters:
        results (ultralytics.yolo.utils.Results): YOLO predictions object.
        num_images (int): Number of images to display.
        conf_threshold (float): Confidence threshold to filter predictions.
    """
    filtered_images = 0
    plt.figure(figsize=(15, 15))

    for i, result in enumerate(results):
        if filtered_images >= num_images:
            break

        # Extract predictions with high confidence
        boxes = result.boxes.xyxy  # Bounding boxes
        confs = result.boxes.conf  # Confidence scores
        labels = result.boxes.cls  # Class IDs

        # Filter predictions by confidence threshold
        high_conf_idx = [idx for idx, conf in enumerate(confs) if conf >= conf_threshold]

        if len(high_conf_idx) > 0:
            img_path = result.path
            img = cv2.imread(img_path)
            img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)  # Convert to RGB for plotting

            # Plot the image and overlay bounding boxes
            plt.subplot(1, num_images, filtered_images + 1)
            plt.imshow(img)
            plt.axis("off")
            plt.title(f"Image: {img_path.split('/')[-1]}")

            for idx in high_conf_idx:
                box = boxes[idx].cpu().numpy()
                conf = confs[idx].item()
                label = int(labels[idx].item())
                x1, y1, x2, y2 = map(int, box)

                # Overlay bounding boxes and confidence score
                plt.gca().add_patch(plt.Rectangle((x1, y1), x2 - x1, y2 - y1,
                                                  edgecolor="green", linewidth=2, fill=False))
                plt.text(x1, y1 - 5, f"{label} ({conf:.2f})", color="green", fontsize=10, bbox=dict(facecolor='white', alpha=0.5))

            filtered_images += 1

    plt.tight_layout()
    plt.show()

# Call the function to visualize positive predictions
visualize_positive_predictions(results, num_images=5, conf_threshold=0.7)
```

```python
def visualize_false_positives(results, annotations_dir, num_images=5, conf_threshold=0.25):
    """
    Visualize false positives from the test set predictions.

    Parameters:
        results (ultralytics.yolo.utils.Results): YOLO predictions object.
        annotations_dir (str): Directory containing ground truth annotations.
        num_images (int): Number of images to display.
        conf_threshold (float): Confidence threshold to filter predictions.
    """
    filtered_images = 0
    plt.figure(figsize=(15, 15))

    for i, result in enumerate(results):
        if filtered_images >= num_images:
            break

        # Extract predictions
        boxes = result.boxes.xyxy  # Predicted bounding boxes
        confs = result.boxes.conf  # Confidence scores
        labels = result.boxes.cls  # Class IDs
        img_path = result.path

        # Load ground truth annotations
        annotation_path = os.path.join(annotations_dir, os.path.basename(img_path).replace(".jpg", ".txt").replace(".png", ".txt"))
        if not os.path.exists(annotation_path):
            print(f"Warning: Annotation file not found for {img_path}")
            continue

        with open(annotation_path, "r") as f:
            gt_boxes = []
            for line in f:
                parts = line.strip().split()
                x_center, y_center, width, height = map(float, parts[1:5])
                x1 = int((x_center - width / 2) * result.orig_shape[1])
                y1 = int((y_center - height / 2) * result.orig_shape[0])
                x2 = int((x_center + width / 2) * result.orig_shape[1])
                y2 = int((y_center + height / 2) * result.orig_shape[0])
                gt_boxes.append([x1, y1, x2, y2])

        # Identify false positives
        false_positives = []
        for idx, box in enumerate(boxes):
            if confs[idx] < conf_threshold:
                continue
            box_np = box.cpu().numpy()
            is_fp = True
            for gt_box in gt_boxes:
                iou = calculate_iou(box_np, gt_box)
                if iou > 0.5:  # Consider it a match if IoU > 0.5
                    is_fp = False
                    break
            if is_fp:
                false_positives.append((box_np, confs[idx].item(), labels[idx].item()))

        if len(false_positives) > 0:
            img = cv2.imread(img_path)
            img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

            plt.subplot(1, num_images, filtered_images + 1)
            plt.imshow(img)
            plt.axis("off")
            plt.title(f"Image: {os.path.basename(img_path)}")

            for box, conf, label in false_positives:
                x1, y1, x2, y2 = map(int, box)
                plt.gca().add_patch(plt.Rectangle((x1, y1), x2 - x1, y2 - y1,
                                                  edgecolor="red", linewidth=2, fill=False))
                plt.text(x1, y1 - 5, f"FP {label} ({conf:.2f})", color="red", fontsize=10, bbox=dict(facecolor='white', alpha=0.5))

            filtered_images += 1

    plt.tight_layout()
    plt.show()

def calculate_iou(box1, box2):
    """
    Calculate Intersection over Union (IoU) for two bounding boxes.

    Parameters:
        box1 (list or numpy array): [x1, y1, x2, y2] coordinates of the first box.
        box2 (list or numpy array): [x1, y1, x2, y2] coordinates of the second box.

    Returns:
        float: IoU value.
    """
    x1 = max(box1[0], box2[0])
    y1 = max(box1[1], box2[1])
    x2 = min(box1[2], box2[2])
    y2 = min(box1[3], box2[3])

    intersection = max(0, x2 - x1 + 1) * max(0, y2 - y1 + 1)
    box1_area = (box1[2] - box1[0] + 1) * (box1[3] - box1[1] + 1)
    box2_area = (box2[2] - box2[0] + 1) * (box2[3] - box2[1] + 1)
    union = box1_area + box2_area - intersection

    return intersection / union

# Call the function to visualize false positives
visualize_false_positives(results, annotations_dir="./datasets/ExDark_Preprocessed/labels/test", num_images=5, conf_threshold=0.25)
```

```python
def visualize_false_negatives(results, annotations_dir, num_images=5, conf_threshold=0.25):
    """
    Visualize false negatives from the test set predictions.

    Parameters:
        results (ultralytics.yolo.utils.Results): YOLO predictions object.
        annotations_dir (str): Directory containing ground truth annotations.
        num_images (int): Number of images to display.
        conf_threshold (float): Confidence threshold to filter predictions.
    """
    filtered_images = 0
    plt.figure(figsize=(15, 15))

    for i, result in enumerate(results):
        if filtered_images >= num_images:
            break

        # Extract predictions
        boxes = result.boxes.xyxy  # Predicted bounding boxes
        confs = result.boxes.conf  # Confidence scores
        labels = result.boxes.cls  # Class IDs
        img_path = result.path

        # Load ground truth annotations
        annotation_path = os.path.join(annotations_dir, os.path.basename(img_path).replace(".jpg", ".txt").replace(".png", ".txt"))
        if not os.path.exists(annotation_path):
            print(f"Warning: Annotation file not found for {img_path}")
            continue

        with open(annotation_path, "r") as f:
            gt_boxes = []
            for line in f:
                parts = line.strip().split()
                class_id = int(parts[0])
                x_center, y_center, width, height = map(float, parts[1:5])
                x1 = int((x_center - width / 2) * result.orig_shape[1])
                y1 = int((y_center - height / 2) * result.orig_shape[0])
                x2 = int((x_center + width / 2) * result.orig_shape[1])
                y2 = int((y_center + height / 2) * result.orig_shape[0])
                gt_boxes.append((class_id, [x1, y1, x2, y2]))

        # Identify false negatives
        false_negatives = []
        for class_id, gt_box in gt_boxes:
            is_fn = True
            for idx, box in enumerate(boxes):
                if confs[idx] < conf_threshold:
                    continue
                pred_box = box.cpu().numpy()
                iou = calculate_iou(pred_box, gt_box)
                if iou > 0.5:  # Consider it a match if IoU > 0.5
                    is_fn = False
                    break
            if is_fn:
                false_negatives.append((class_id, gt_box))

        if len(false_negatives) > 0:
            img = cv2.imread(img_path)
            img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

            plt.subplot(1, num_images, filtered_images + 1)
            plt.imshow(img)
            plt.axis("off")
            plt.title(f"Image: {os.path.basename(img_path)}")

            # Draw ground truth bounding boxes (false negatives)
            for class_id, box in false_negatives:
                x1, y1, x2, y2 = box
                plt.gca().add_patch(plt.Rectangle((x1, y1), x2 - x1, y2 - y1,
                                                  edgecolor="blue", linewidth=2, fill=False))
                plt.text(x1, y1 - 5, f"FN {class_id}", color="blue", fontsize=10, bbox=dict(facecolor='white', alpha=0.5))

            # Draw predicted bounding boxes
            for idx, box in enumerate(boxes):
                if confs[idx] < conf_threshold:
                    continue
                x1, y1, x2, y2 = map(int, box)
                plt.gca().add_patch(plt.Rectangle((x1, y1), x2 - x1, y2 - y1,
                                                  edgecolor="green", linewidth=2, fill=False))
                plt.text(x1, y1 - 5, f"Pred", color="green", fontsize=10, bbox=dict(facecolor='white', alpha=0.5))

            filtered_images += 1

    plt.tight_layout()
    plt.show()

# Esegui la funzione per visualizzare i falsi negativi
visualize_false_negatives(
    results=results,  # Risultati delle predizioni ottenuti dal modello YOLO
    annotations_dir="./datasets/ExDark_Preprocessed/labels/test",  # Directory delle annotazioni
    num_images=5,  # Numero di immagini da visualizzare
    conf_threshold=0.25  # Confidence threshold per filtrare i falsi negativi
)
```

```python
import os
import shutil
from tqdm import tqdm

# Define paths
image_dir = "./datasets/ExDark_Preprocessed/images/test"
label_dir = "./datasets/ExDark_Preprocessed/labels/test"
output_dir_images = "./datasets/ExDark_Preprocessed/images"
output_dir_labels = "./datasets/ExDark_Preprocessed/labels"

# Filter metadata for test set only
test_metadata = metadata[metadata["TrainValTest"] == 3]

# Trova tutte le categorie di "LightCondition" uniche
light_conditions = test_metadata["LightCondition"].unique()

# Crea directory per ogni categoria di luce
for condition in light_conditions:
    condition_image_dir = os.path.join(output_dir_images, f"test_{condition}")
    condition_label_dir = os.path.join(output_dir_labels, f"test_{condition}")
    os.makedirs(condition_image_dir, exist_ok=True)
    os.makedirs(condition_label_dir, exist_ok=True)

# Dividi le immagini e le annotazioni in base alla tipologia di luce
for _, row in tqdm(test_metadata.iterrows(), desc="Splitting images by light condition", total=len(test_metadata)):
    filename = row["Filename"]
    light_condition = row["LightCondition"]

    # Percorsi sorgente per immagine e label
    src_image_path = os.path.join(image_dir, filename)
    src_label_path = os.path.join(label_dir, f"{os.path.splitext(filename)[0]}.txt")

    # Verifica se i file esistono
    if not os.path.exists(src_image_path):
        filename_lower = filename.lower()
        src_image_path = os.path.join(image_dir, filename_lower)
    if not os.path.exists(src_image_path):
        filename_upper = filename.upper()
        src_image_path = os.path.join(image_dir, filename_upper)

    if not os.path.exists(src_image_path):
        print(f"Warning: Image not found: {src_image_path}")
        continue

    if not os.path.exists(src_label_path):
        print(f"Warning: Label not found: {src_label_path}")
        continue

    # Copia immagine e annotazione nella directory corrispondente
    condition_image_dir = os.path.join(output_dir_images, f"test_{light_condition}")
    condition_label_dir = os.path.join(output_dir_labels, f"test_{light_condition}")
    shutil.copy2(src_image_path, condition_image_dir)
    shutil.copy2(src_label_path, condition_label_dir)
```

```python
def create_yaml_for_light_conditions(output_dir, base_dir, light_conditions, classes):
    """
    Create YAML files for each light condition.

    Parameters:
        output_dir (str): Directory to save YAML files.
        base_dir (str): Base directory for the dataset (e.g., ExDark_Preprocessed).
        light_conditions (list): List of light condition categories.
        classes (list): List of class names.
    """
    for condition in light_conditions:
        yaml_data = {
            "path": base_dir,
            "train": "",  # Placeholder (not used)
            "val": "",  # Placeholder (not used)
            "test": f"images/test_{condition}",  # Test set for the specific light condition
            "names": classes
        }
        yaml_path = os.path.join(output_dir, f"exdark_preprocessed_{condition}.yaml")
        with open(yaml_path, "w") as yaml_file:
            yaml.dump(yaml_data, yaml_file, default_flow_style=False)
        print(f"YAML file created for {condition}: {yaml_path}")

# Define paths and classes
output_yaml_dir = "."
classes = ["Bicycle", "Boat", "Bottle", "Bus", "Car", "Cat",
           "Chair", "Cup", "Dog", "Motorbike", "People", "Table"]

# Generate YAML files for each light condition
create_yaml_for_light_conditions(output_yaml_dir, "./ExDark_Preprocessed", light_conditions, classes)
```

```python
# Dizionario per salvare i risultati
light_condition_metrics = {}

for condition in light_conditions:
    yaml_path = f"./exdark_preprocessed_{condition}.yaml"
    print(f"Evaluating for light condition: {condition}")

    # Calcola le metriche per la condizione corrente
    metrics = model.val(data=yaml_path, split="test")

    # Salva solo il mAP@0.5 nel dizionario
    light_condition_metrics[condition] = metrics.box.map50  # Salva solo il mAP@0.5
```

```python
# Stampa solo il mAP@0.5 per ciascuna condizione
for condition, map50 in light_condition_metrics.items():
    print(f"{condition}: mAP@0.5 = {map50:.4f}")
```
